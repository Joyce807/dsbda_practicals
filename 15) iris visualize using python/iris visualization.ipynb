```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris_df = pd.read_csv('iris.csv')
```

```python
sns.histplot(data=iris_df, x="sepal_length", kde=True)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
sns.histplot(data=iris_df, x="sepal_width", kde=True)
plt.title("Histogram of Sepal Width")
plt.show()
```

```python
sns.histplot(data=iris_df, x="petal_length", kde=True)
plt.title("Histogram of Petal Length")
plt.show()
```

```python
sns.histplot(data=iris_df, x="petal_width", kde=True)
plt.title("Histogram of Petal Width")
plt.show()
```

```python
sns.scatterplot(data=iris_df, x="sepal_length", y="sepal_width", hue="species")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.scatterplot(data=iris_df, x="petal_length", y="petal_width", hue="species")
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
```

```python
sns.boxplot(data=iris_df, y="sepal_length", x="species")
plt.title("Box Plot: Sepal Length")
plt.show()
```

```python
sns.boxplot(data=iris_df, y="sepal_width", x="species")
plt.title("Box Plot: Sepal Width")
plt.show()
```

```python
sns.boxplot(data=iris_df, y="petal_length", x="species")
plt.title("Box Plot: Petal Length")
plt.show()
```

```python
sns.boxplot(data=iris_df, y="petal_width", x="species")
plt.title("Box Plot: Petal Width")
plt.show()
```

```python
sns.pairplot(data=iris_df, hue="species")
plt.title("Pair Plot")
plt.show()
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',data=iris_df)
sns.lmplot(x='petal_length',y='petal_width',data=iris_df)
```

```python
sns.residplot(x = iris_df['sepal_length'], y =iris_df['sepal_width'], color='red')
sns.residplot(x = iris_df['petal_length'], y =iris_df['petal_width'], color='red')
```

```python
plt.scatter(iris_df['sepal_length'], iris_df['sepal_width'], label='iris', color='red',marker = 'o')
```

```python
sns.heatmap(iris_df.corr(),linewidth=0.3,vmax=1.0,square=True, linecolor='black',annot=True)
```
